```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
from matplotlib.ticker import MaxNLocator
```

```python
def plot_forecast(train, test, fitted, forecast, date, title):
    plt.figure(figsize=(14, 6))

    date_train = date[:len(train)]
    date_test = date[-len(test):]

    plt.plot(date_train, train, label='Train', color='blue', alpha=0.6)
    plt.plot(date_test, test, label='Actual', color='green', linewidth=1.5)
    plt.plot(date_train, fitted, label='Fitted', color='orange', linestyle='--', linewidth=1.5)
    plt.plot(date_test, forecast, label='Forecast', color='red', linewidth=2)
    plt.xlabel('Дата')
    plt.gca().xaxis.set_major_locator(MaxNLocator(nbins=12))
    plt.xticks(rotation=45)

    plt.title(title)
    plt.ylabel('Значение')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def print_metrics(forecast, test):
    mae = mean_absolute_error(test, forecast)
    rmse = np.sqrt(mean_squared_error(test, forecast))
    print(f"\nMAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
```

```python
data = pd.read_csv("../data/processed/ZK 68, (48-1, 48-air), 27.11.20-15.12.25.csv")
X = data
```

```python

def cv(train, seasonal_periods, trend, season):
    fold_size = seasonal_periods
    train_end_index = fold_size * 2 + 1
    sum_mse = 0
    iter = 0
    while train_end_index + fold_size < len(train_shifted):
        es_cv = ExponentialSmoothing(
            train[:train_end_index],
            seasonal_periods=seasonal_periods,
            trend=trend,
            seasonal=season,
            initialization_method="heuristic"
        ).fit()
        pred = es_cv.forecast(seasonal_periods)
        sum_mse += mean_squared_error(train_shifted[train_end_index:train_end_index + fold_size], pred)
        iter += 1
        train_end_index += fold_size
    return sum_mse / iter
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
def grid_search(train_shifted):
    best_score = 10e9
    best_params = None
    seasonal_periods = 365
    for trend in ["add", "mul", None]:
        for season in ["add", "mul", None]:
            mse_cv = cv(train_shifted, seasonal_periods, trend, season)
            if mse_cv < best_score:
                best_params = {"trend": trend, "season": season}
                best_score = mse_cv

    print(f"\nBest params: {best_params} ")
    return ExponentialSmoothing(
                train_shifted,
                seasonal_periods=365,
                trend=best_params["trend"],
                seasonal=best_params["season"],
                initialization_method="heuristic",
            ).fit()
```

```python

date = pd.to_datetime(data['Дата'])
test_size = len(data[data['Год'] == 2025])
train_size = len(data) - test_size
for hole in ['48-Воздух (0)', '48-1 (3)', '48-1 (5)', '48-1 (10)']:
    shift = np.abs(np.min(X[hole][:train_size])) + 1
    train, test = X[hole][:train_size], X[hole][train_size:]
    train_shifted = train + shift
    es = grid_search(train_shifted)
    fitted = es.fittedvalues - shift
    forecast = es.forecast(test_size) - shift
    plot_forecast(train, test, fitted, forecast, date, hole)
    print_metrics(forecast, test)
```
